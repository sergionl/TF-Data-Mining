## To transactionnal

```python
def to_transactionnal(df, column_trans, column_items):
  transactions = []
  for v in df[column_trans].unique():
    transactions.append(list(df[df[column_trans] == v][column_items].values))
  return transactions

trans = to_transactionnal(df_siembra, 'UBICACION', 'ESPECIE')
print(len(trans))
```

```python
print(df_siembra['UBICACION'].unique()[0], trans[0])
print(df_siembra['UBICACION'].unique()[20], trans[20])
print(df_siembra['UBICACION'].unique()[50], trans[50])
print(df_siembra['UBICACION'].unique()[120], trans[120])
print(df_siembra['UBICACION'].unique()[500], trans[500])
```

```python
r = fpgrowth(trans, target='r', supp=10, zmin=3,conf=50,report='scl')

```

```python
pd.set_option('display.max_colwidth', None)
df_r = pd.DataFrame(r)
df_r.columns = ['Consecuente', 'Antecendente', 'Freq', 'Freq(%)', 'Conf']
df_r.sort_values(by='Conf', ascending=False).head(10)
```
